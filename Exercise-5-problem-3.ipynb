### Part 3 (2 points)

The median temperatures above consider data from the entire "summer" (Mar-May), hence the differences might not be so clear. Let's now find out the mean temperatures from March and May 2018 in Mactan and Hinatuan.

- From the `mactan` and `hinatuan` DataFrames, select the rows where values of the `DATE` column are from March 2018
    - Assign these selected rows into the variables `mactan_mar` and `hinatuan_mar` (you can check the [hints](https://geo-python.github.io/site/lessons/L5/exercise-5.html) for help!)
- Repeat the procedure for the month of May and assign those values into variables to `mactan_may` and `hinatuan_may`

```python
# Select the subset of the Mactan and Hinatuan data for the 3rd and 5th month
mactan_mar = None
hinatuan_mar = None

mactan_may = None
hinatuan_may = None

# YOUR CODE HERE
# The data from March and May in both mactan and hinatuan were assigned to their respective variables
# Unfortunately, we encountered some errors with the symbols for unknown reason, thus we opted to separate the data from March and May
mactan_mar = mactan.loc[mactan['DATE'] >= '2018-03-01']
mactan_mar = mactan.loc[mactan['DATE'] <= '2018-03-31']
hinatuan_mar = hinatuan.loc[hinatuan['DATE'] >= '2018-03-01']
hinatuan_mar = hinatuan.loc[hinatuan['DATE'] <= '2018-03-31']

mactan_may = mactan.loc[mactan['DATE'] >= '2018-05-01']
mactan_may = mactan.loc[mactan['DATE'] <= '2018-05-31']
hinatuan_may = hinatuan.loc[hinatuan['DATE'] >= '2018-05-01']
hinatuan_may = hinatuan.loc[hinatuan['DATE'] <= '2018-05-31']
```

### Part 4 (1 point)

Now you can make your temperature data from both locations and months easier to compare by printing out a few useful values.

- Use the `print()` function to show the mean, min and max Celsius temperatures for both places in March and May using the new subset dataframes (`mactan_mar`, `hinatuan_mar`, `mactan_may`, and `hinatuan_may`).

```python
# YOUR CODE HERE
# mean(), min(), and max() functions were used to determine the mean, minimum, and maximum values in the data
mactan_mar_mean = mactan_mar["Celsius"].mean()
hinatuan_mar_mean = hinatuan_mar["Celsius"].mean()
mactan_may_mean = mactan_may["Celsius"].mean()
hinatuan_may_mean = hinatuan_may["Celsius"].mean()

mactan_mar_min = mactan_mar["Celsius"].min()
hinatuan_mar_min = hinatuan_mar["Celsius"].min()
mactan_may_min = mactan_may["Celsius"].min()
hinatuan_may_min = hinatuan_may["Celsius"].min()

mactan_mar_max = mactan_mar["Celsius"].max()
hinatuan_mar_max = hinatuan_mar["Celsius"].max()
mactan_may_max = mactan_may["Celsius"].max()
hinatuan_may_max = hinatuan_may["Celsius"].max()

# For a better presentation of the result, we assigned the values to another variable and created a conclusion paragraph to state the results
print (f"The means of mactan {mactan_mar_mean} in March and {mactan_may_mean} in May, while the means of hinatuan are {hinatuan_mar_mean} in March and {hinatuan_may_mean} in May. The min of mactan are {mactan_mar_min} in March and {mactan_may_min} in May, while the min of hinatuan are {hinatuan_mar_min} in March and {hinatuan_may_min} in May. The max of mactan are {mactan_mar_max} in March and {mactan_may_max} in May, while the max of hinatuan are {hinatuan_mar_max} in March and {hinatuan_may_max} in May.")
```
